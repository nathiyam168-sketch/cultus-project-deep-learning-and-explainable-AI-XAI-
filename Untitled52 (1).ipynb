```python
"""
Advanced_TimeSeries_Forecasting_XAI.py

Full end-to-end project template that:
- Creates or loads a multivariate time series dataset (synthetic electricity v/s weather example included)
- Preprocesses (scaling, windowing) for multi-step forecasting
- Builds and trains TWO model types: LSTM and a simple TCN (Conv1D dilation-based)
- Performs hyperparameter-friendly training pipeline
- Evaluates with RMSE, MAE, MAPE
- Applies XAI using SHAP (with fallback to permutation importance if SHAP unavailable)
- Produces plots and a ready-to-submit report (Markdown)

Run this file in a Python environment with the listed dependencies.

Dependencies (pip):
tensorflow, numpy, pandas, scikit-learn, matplotlib, shap (optional), tqdm

If running in Colab: install missing packages with pip before executing.

"""

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers
import json
import datetime

# Optional imports (SHAP)
try:
    import shap
    SHAP_AVAILABLE = True
except Exception:
    SHAP_AVAILABLE = False

# ----------------------
# SETTINGS
# ----------------------
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# User options
USE_SYNTHETIC = True   # If False, set CSV path below to load a dataset
CSV_PATH = ""        # e.g. './electricity_weather.csv'
TARGET_COL = 'target'  # name used for synthetic dataset
N_INPUT_FEATURES = 3
MULTI_STEP = 6         # predict next 6 time steps
LOOKBACK = 48          # number of past timesteps to feed model
BATCH_SIZE = 32
EPOCHS = 25
MODEL_SAVE_DIR = 'saved_models'
os.makedirs(MODEL_SAVE_DIR, exist_ok=True)

# ----------------------
# 1) DATASET (synthetic) - electricity consumption influenced by temperature, humidity
# ----------------------

def generate_synthetic_timeseries(n_timesteps=5000, seed=RANDOM_SEED):
    np.random.seed(seed)
    t = np.arange(n_timesteps)
    # seasonal daily cycle and weekly variation
    daily = 10 * np.sin(2 * np.pi * t / 24)
    weekly = 5 * np.sin(2 * np.pi * t / (24*7))
    trend = 0.001 * t

    # features
    temperature = 20 + 8 * np.sin(2 * np.pi * t / 24) + 2 * np.sin(2 * np.pi * t / (24*7)) + np.random.normal(0,0.5,n_timesteps)
    humidity = 60 + 10 * np.cos(2 * np.pi * t / 24) + np.random.normal(0,1,n_timesteps)
    other = 2 * np.sin(2 * np.pi * t / 12) + np.random.normal(0,0.5,n_timesteps)

    # target: electricity consumption depends on temperature (AC usage) and humidity slightly
    base_consumption = 50 + daily + weekly + trend*5
    target = base_consumption + 0.8 * np.maximum(0, temperature - 22) + 0.05 * humidity + np.random.normal(0,1.5,n_timesteps)

    df = pd.DataFrame({
        'temperature': temperature,
        'humidity': humidity,
        'other': other,
        'target': target
    })
    return df

# Load or generate
if USE_SYNTHETIC or (not CSV_PATH):
    df = generate_synthetic_timeseries(n_timesteps=6000)
else:
    df = pd.read_csv(CSV_PATH)
    # user must ensure df contains necessary columns

print('Data shape:', df.shape)
print(df.head())
```

```python
# ----------------------
# 2) Preprocessing: scaling and windowing for multi-step forecasting
# ----------------------

def create_windows(data, lookback=48, horizon=6, target_col='target'):
    X, Y = [], []
    values = data.values
    n = len(data)
    for i in range(lookback, n - horizon + 1):
        X.append(values[i-lookback:i, :])
        Y.append(values[i:i+horizon, data.columns.get_loc(target_col)])
    X = np.array(X)
    Y = np.array(Y)
    return X, Y
# scale
scaler = StandardScaler()
feature_cols = df.columns.tolist()  # including target
scaled = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled, columns=feature_cols)

X, Y = create_windows(scaled_df, lookback=LOOKBACK, horizon=MULTI_STEP, target_col=TARGET_COL)
print('X shape, Y shape:', X.shape, Y.shape)

# train/val/test split
train_size = int(0.7 * len(X))
val_size = int(0.15 * len(X))

X_train, Y_train = X[:train_size], Y[:train_size]
X_val, Y_val = X[train_size:train_size+val_size], Y[train_size:train_size+val_size]
X_test, Y_test = X[train_size+val_size:], Y[train_size+val_size:]

print('Train/Val/Test sizes:', X_train.shape[0], X_val.shape[0], X_test.shape[0])
```

```python
# ----------------------
# 3) Model builders
# ----------------------

def build_lstm_model(input_shape, horizon, units=64, dropout=0.2):
    inp = layers.Input(shape=input_shape)
    x = layers.LSTM(units, return_sequences=False)(inp)
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(units//2, activation='relu')(x)
    out = layers.Dense(horizon)(x)  # multi-step direct output
    model = models.Model(inp, out)
    model.compile(optimizer='adam', loss='mse')
    return model

def build_tcn_model(input_shape, horizon, filters=64, kernel_size=3, dilation_rates=[1,2,4], dropout=0.2):
    inp = layers.Input(shape=input_shape)
    x = inp
    for d in dilation_rates:
        x = layers.Conv1D(filters, kernel_size, dilation_rate=d, padding='causal', activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(dropout)(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(filters//2, activation='relu')(x)
    out = layers.Dense(horizon)(x)
    model = models.Model(inp, out)
    model.compile(optimizer='adam', loss='mse')
    return model
    print('\nLSTM metrics:', metrics_lstm)
    print('TCN metrics:', metrics_tcn)
```

```python
# ----------------------
# 4) Train both models and evaluate
# ----------------------

input_shape = X_train.shape[1:]  # (lookback, n_features)
horizon = Y_train.shape[1]

lstm = build_lstm_model(input_shape, horizon, units=128)
tcn = build_tcn_model(input_shape, horizon, filters=64)

print('LSTM summary:')
lstm.summary()
print('\nTCN summary:')
tcn.summary()

es = callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history_lstm = lstm.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=EPOCHS, batch_size=BATCH_SIZE, callbacks=[es], verbose=2)
model_path_lstm = os.path.join(MODEL_SAVE_DIR, 'lstm_model.h5')
lstm.save(model_path_lstm)

history_tcn = tcn.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=EPOCHS, batch_size=BATCH_SIZE, callbacks=[es], verbose=2)
model_path_tcn = os.path.join(MODEL_SAVE_DIR, 'tcn_model.h5')
tcn.save(model_path_tcn)
```

```python
# ----------------------
# 5) Evaluation functions
# ----------------------

def evaluate_model(model, X_test, Y_test, scaler, df_columns, horizon):
    preds = model.predict(X_test)
    # preds and Y_test are scaled values for target - we need to invert scaling only for target
    # scaler works on full feature vector; we'll reconstruct arrays to inverse-transform target only
    def inverse_target(series_scaled_values):
# series_scaled_values: shape (n_samples, horizon)
        n = series_scaled_values.shape[0]
        out = []
        for i in range(horizon):
# take column i of horizon predictions and inverse transform by constructing a dummy feature matrix
            col = series_scaled_values[:, i]
            dummy = np.zeros((len(col), len(df_columns)))
# set target's scaled column = col
            dummy[:, df_columns.index(TARGET_COL)] = col
            inv = scaler.inverse_transform(dummy)[:, df_columns.index(TARGET_COL)]
            out.append(inv)
# stack -> shape (n_samples, horizon)
        return np.vstack(out).T

    preds_inv = inverse_target(preds)
    y_inv = inverse_target(Y_test)
# compute metrics averaged across all horizon steps
    rmse = np.sqrt(mean_squared_error(y_inv.flatten(), preds_inv.flatten()))
    mae = mean_absolute_error(y_inv.flatten(), preds_inv.flatten())
    mape = np.mean(np.abs((y_inv - preds_inv) / (y_inv + 1e-8))) * 100
    return {'rmse': rmse, 'mae': mae, 'mape': mape}, preds_inv, y_inv

metrics_lstm, preds_lstm, y_test_inv = evaluate_model(lstm, X_test, Y_test, scaler, feature_cols, horizon)
metrics_tcn, preds_tcn, _ = evaluate_model(tcn, X_test, Y_test, scaler, feature_cols, horizon)
```

```python
# ----------------------
# 6) Plots: Actual vs Predicted for first test samples
# ----------------------

def plot_sample_predictions(preds_inv, y_inv, sample_idx=0, horizon=6):
    plt.figure(figsize=(10,4))
    plt.plot(np.arange(horizon), y_inv[sample_idx], label='Actual')
    plt.plot(np.arange(horizon), preds_inv[sample_idx], label='Predicted')
    plt.xlabel('Horizon step')
    plt.ylabel('Target')
    plt.title(f'Sample {sample_idx} - Actual vs Predicted')
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_sample_predictions(preds_lstm, y_test_inv, sample_idx=5, horizon=horizon)
plot_sample_predictions(preds_tcn, y_test_inv, sample_idx=5, horizon=horizon)
```

```python
# ----------------------
# 7) XAI: SHAP explanation for model predictions (or fallback permutation importance)
# ----------------------

if SHAP_AVAILABLE:
    print('\nRunning SHAP explanations (may take time) ...')
    # For deep TF models, GradientExplainer or DeepExplainer can be used depending on model.
    explainer = shap.GradientExplainer(lstm, X_train[:200])
    # explain a small set
    shap_values = explainer.shap_values(X_test[:200])
    # shap_values shape depends on explainer; for regression multicontinuous output, handle first horizon component
    try:
        # shap_values for outputs -> list, take first output
        sv = shap_values[0]
    except Exception:
        sv = shap_values
    # aggregate feature importance by averaging absolute shap values over timesteps and samples
    # X_test shape: (n, lookback, features)
    mean_abs_shap = np.mean(np.abs(sv), axis=(0,1))  # mean over samples & time
    # plot
    plt.figure(figsize=(8,3))
    plt.bar(feature_cols, mean_abs_shap)
    plt.title('Mean |SHAP| by feature (aggregated over lookback/time)')
    plt.tight_layout()
    plt.show()
else:
    print('\nSHAP not available. Performing permutation importance (approx) ...')
    # baseline error
    baseline_preds = lstm.predict(X_test)
    def compute_rmse_flat(preds, y_true):
        return np.sqrt(mean_squared_error(y_true.flatten(), preds.flatten()))
    baseline_rmse = compute_rmse_flat(baseline_preds, Y_test)
    import copy
    perm_importance = {}
    for feat_idx, feat_name in enumerate(feature_cols):
        X_perm = X_test.copy()
        # shuffle the feature across samples and time
        shuffled = X_perm[:,:,feat_idx].flatten()
        np.random.shuffle(shuffled)
        X_perm[:,:,feat_idx] = shuffled.reshape(X_perm.shape[0], X_perm.shape[1])
        preds_perm = lstm.predict(X_perm)
        rmse_perm, _, _ = evaluate_model(lstm, X_perm, Y_test, scaler, feature_cols, horizon)
        perm_importance[feat_name] = rmse_perm['rmse'] - baseline_rmse
    # plot
    names = list(perm_importance.keys())
    values = [perm_importance[n] for n in names]
    plt.figure(figsize=(8,3))
    plt.bar(names, values)
    plt.title('Permutation importance (increase in RMSE when feature permuted)')
    plt.tight_layout()
    plt.show()
```

```python
# ----------------------
# 8) Save a short markdown report summarizing results
# ----------------------

report = {
    'project_title': 'Advanced Time Series Forecasting with Deep Learning + XAI',
    'date': str(datetime.date.today()),
    'dataset': 'synthetic electricity vs weather' if USE_SYNTHETIC else CSV_PATH,
    'n_samples': len(df),
    'features': feature_cols,
    'lookback': LOOKBACK,
    'horizon': MULTI_STEP,
    'metrics': {
        'lstm': metrics_lstm,
        'tcn': metrics_tcn
    },
    'model_paths': {
        'lstm': model_path_lstm,
        'tcn': model_path_tcn
    },
    'notes': 'SHAP used for XAI' if SHAP_AVAILABLE else 'Permutation importance used as fallback'
}
report_md = f"""# Project Report - {report['project_title']}

*Date:* {report['date']}

*Dataset:* {report['dataset']} ({report['n_samples']} rows)

*Features:* {', '.join(report['features'])}

*Modeling setup*
- Lookback: {report['lookback']}
- Forecast horizon (multi-step): {report['horizon']}

*Results*
- LSTM: RMSE={metrics_lstm['rmse']:.4f}, MAE={metrics_lstm['mae']:.4f}, MAPE={metrics_lstm['mape']:.2f}%
- TCN:  RMSE={metrics_tcn['rmse']:.4f}, MAE={metrics_tcn['mae']:.4f}, MAPE={metrics_tcn['mape']:.2f}%

*XAI*
- {report['notes']}

*Files saved*
- {model_path_lstm}
- {model_path_tcn}

"""
with open('report.md', 'w') as f:
    f.write(report_md)

print('\nReport saved to report.md')
```
